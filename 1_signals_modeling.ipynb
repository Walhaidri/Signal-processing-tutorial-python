```python
"""
Created on Tue Sep 20 14:53:19 2022

@author: waleed.al.haidri
"""
import numpy as np
from numpy import arange, sin, cos, pi
import  matplotlib.pyplot as plt

# Simulation of deterministic periodic harmonic signals
# Study signal parameters

rad = np.arange(-pi,pi,0.1) # list of phase in radian
deg = [round(i *180/pi, 3) for i in rad] # list of phase in degrees
a = np.arange(1,10) # list of Amplitudes
f =np.arange(1,10) #  list of frequencies
fs =100 # sample rate
T = 1/fs # sample time
sec = 1 # duration
t = arange(0,sec,T) # time
for i in range(len(a)):
    y = a[0]*sin(2*pi*f[i]*t + deg[0]) # 
    # plot model
    plt.figure(i)
    plt.plot(t,y, '-*')
    plt.title( 'frequency = ' + str(f[i]))
    plt.xlabel('time, sec')
    plt.ylabel('amplitude, V')
    plt.show()
    plt.pause(0.5)
```

```python
#%% Modeling Deterministic Periodic Polyharomic Signals

a = [5,3,4,7] # list of amplitudes
f = [0,2,4,6] # list of frquencies
ph = [0, -0.4, -0.6, -0.8] # list of phases 
fs =50 # sampling rate
T = 1/fs # sample time
sec = 5 # duration
t = np.arange(0,sec,T) # time axis
z=0
for i in range(len(a)):
    z += a[i]*sin(2*pi*f[i]*t + ph[i]) # model

    # plot model
    plt.plot(t,z)
    plt.title('Polyharomic Signal')
    plt.xlabel('time, sec')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
#%% Modeling Deterministic non Periodic Polyharomic Signals

a = [5,3,4,7] # list of amplitudes
f = [0,2,np.sqrt(5),7]  # list of frquencies
ph = [0, -0.4, -0.6, -0.8] # list of phases 
fs =50 # sampling rate
T = 1/fs # sample time
sec = 5 # duration
t = np.arange(0,sec,T) # time axis
z=0
for i in range(len(a)):
    z += a[i]*sin(2*pi*f[i]*t + ph[i]) # model
    # plot model
    plt.plot(t,z)
    plt.title('Polyharomic Signal')
    plt.xlabel('time, sec')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
#%% Modeling deterministic aperiodic signals

a = 0.15 # 
b = 0.17 # 
T = 1/fs # sample time
sec = 100 # duration
t = arange(0,sec,T) # time
s =np.exp(-a*t) - np.exp(-b*t) # model

# plot model
plt.figure(2)
plt.plot(t,s)
plt.title('Aperiodic signals')
plt.xlabel('time, sec')
plt.ylabel('amplitude, V')
plt.show()
```

```python
#%% Simulation of random signals| noise

a = 1 # Amplitude
T = 0.001 # sample time
sec = 1 # duration
t = arange(0,sec,T) # time
noise = a*np.random.rand(len(t))

# plot model
plt.figure(figsize=[15, 8])
plt.subplot(1,2,1)
plt.plot(t,noise)
plt.subplot(1,2,2)
plt.hist(noise)
plt.show()
```
